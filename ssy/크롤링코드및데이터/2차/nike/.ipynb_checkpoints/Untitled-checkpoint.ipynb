```python
import pandas as pd
filename = './nike_W_B.csv'
```

```python
df = pd.read_csv(filename)
df = df.drop(columns=['prod_id'])
df['0'] = 'nike'
df.rename(columns={'href':'1',
                  'price':'2',
                  'img':'3'}, inplace=True)
df = df[['0','1','2','3']]
df.to_csv(f'{filename}')
```

```python
# 가격 
for idx, line in df.iterrows():
    df.loc[idx,'price'] = df.loc[idx,'price'].split('n')[-1]
df
```

```python
# 상품id만 추출
for idx, line in df.iterrows():
#     print(df.loc[idx,'img'].split('/')[7].split('_')[0])
    df.loc[idx,'prod_id'] = df.loc[idx, 'img'].split('/')[7].split('_')[0]
df
```

```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
    
error_id=[]
for idx, line in df.iterrows():
#     if df.loc[idx,'href'] == 0:
    prod_id = df.loc[idx, 'prod_id']
    url = f'https://www.nike.com/kr/ko_kr/search?q={prod_id}'
    res = requests.get(url,headers={'User-Agent': 'Mozilla/5.0'})
    soup = BeautifulSoup(res.content, 'html.parser', from_encoding='utf-8')
    try:
        href = 'https://www.nike.com/' + soup.select_one('div.a-product-image.item-imgwrap.action-hover > a')['href']
    except TypeError:
        try:
            href = 'https://www.nike.com/' + soup.select_one('a.a-product-image-link')['href']
        except:
            print(prod_id)
            error_id.append(prod_id)
            href = 0
    df.loc[idx, 'href'] = href
```

```python
df = df[['href','price','img','prod_id']]
df
```

```python
df.to_csv(f'{filename}')
```
