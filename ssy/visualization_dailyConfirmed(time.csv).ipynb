```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)
```

```python
df_time = pd.read_csv(r'../dataset/Time.csv')

yesterday = 0
today = 0
daily_confirmed=[]

for confirmed in df_time['confirmed']:
    yesterday = today
    today = confirmed
    daily_confirmed.append(today - yesterday)
df_time['daily'] = (daily_confirmed)
```

```python
plt.rcParams["figure.figsize"] = (10,5)
pd.pivot_table(df_time, index='date', values='daily').plot()
plt.title('일일 확진자 수(2020/01/20~05/31)')
plt.xlabel('년-월-일')
plt.ylabel('확진자 수 (명)')
plt.show()
```
